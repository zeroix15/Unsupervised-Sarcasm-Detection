```python
import pandas as pd

# Initialize an empty list to hold the data
data = []

# Loop through each file
for i in range(1, 1001):
    file_path = f"/content/drive/My Drive/AlifResearch/sarcasm_v1/sarc/sarcastic_{i}.txt"

    # Read the file and get its content
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read().strip()
        # Append a tuple (class, text) to the data list
        data.append(('sarc', text))

# Create the DataFrame
df_sarcasm = pd.DataFrame(data, columns=['class', 'text'])

# Display the first few rows of the DataFrame
print(df_sarcasm.head())

# Optionally, save the DataFrame to a CSV file
df_sarcasm.to_csv('/content/drive/My Drive/AlifResearch/sarcasm_v1/sarcasm_v1_sarc.csv', index=False)
```

```python
import pandas as pd
import glob
import os

# Path to the folder containing the .txt files
folder_path = "/content/drive/My Drive/AlifResearch/sarcasm_v1/sarc/"

# Get a list of all .txt files in the folder
txt_files = glob.glob(os.path.join(folder_path, "*.txt"))

# Initialize an empty list to hold the data
data = []

# Loop through each file
for file_path in txt_files:
    # Read the file and get its content
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read().strip()
        # Append a tuple (class, text) to the data list
        data.append(('sarc', text))

# Create the DataFrame
df_sarcasm = pd.DataFrame(data, columns=['class', 'text'])

# Display the first few rows of the DataFrame
print(df_sarcasm.head())

# Optionally, save the DataFrame to a CSV file
df_sarcasm.to_csv('/content/drive/My Drive/AlifResearch/sarcasm_v1/sarcasm_v1_sarc.csv', index=False)
```
